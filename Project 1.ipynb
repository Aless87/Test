```python
Cantons_number = Cantons.shape[0]
print(Cantons_number)
Height_matrice=pd.DataFrame(0,index=Cantons,columns=['Height Difference'])
for canton in Cantons: 
    Canton2=df[df["Canton"] == canton]
    min_height=Canton2['Elevation'].min()
    max_height=Canton2['Elevation'].max()
    height_difference=max_height-min_height
    print(f"height difference in {canton} = {height_difference}")
    Height_matrice.loc[canton, 'Height Difference'] = height_difference
    
Height_matrice.sort_values(by='Height Difference',ascending=False)[:5]
```

```python
#Section C.1
import matplotlib.pyplot as plt
populated=df.sort_values(by='Residents',ascending=False)[:10]
plt.figure(figsize=(15, 12))
plt.barh(populated['Commune'],populated['Residents'])
plt.xlabel("Population")
plt.ylabel("Commune")
plt.title("10 most populated communes")
```

```python
#Section C.2
settlement_area=populated['Settlement area']
agricultural_area=populated['Agricultural area']
wooded_area=populated['Wooded area']
unproductive_area=populated['Unproductive area']
plt.barh(populated['Commune'], settlement_area, label='Settlement area')
plt.barh(populated['Commune'], agricultural_area, left=settlement_area, label='Agricultural area')
plt.barh(populated['Commune'], wooded_area, left=settlement_area + agricultural_area, label='Wooded area')
plt.barh(populated['Commune'], unproductive_area, left=settlement_area + agricultural_area + wooded_area, label='Unproductive area')
plt.xlabel("area type in %")
plt.ylabel("Commune")
plt.title("10 most populated communes by type of area")
plt.legend()
```

```python
#Section C.5
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
sns.boxplot(data=df, x='Language', y='0-19 years', ax=axes[0])
sns.boxplot(data=df, x='Language', y='20-64 years', ax=axes[1])
sns.boxplot(data=df, x='Language', y='65 years or over', ax=axes[2])
axes[0].set_title('Age Group 0-19 Years')
axes[0].set_ylabel('Percentage')
axes[1].set_title('Age Group 20-64 Years')
axes[1].set_ylabel('Percentage')
axes[2].set_title('Age Group 65 years or over')
axes[2].set_ylabel('Percentage')
```

```python
#Section C.6
sns.pairplot(df, vars=['Settlement area', 'Agricultural area', 'Wooded area', 'Unproductive area'])

sns.pairplot(df, x_vars=["Settlement area"], y_vars=["Agricultural area"], hue="Elevation")
plt.title('Agricultural area vs. Settlement area')
plt.show()

sns.pairplot(df, x_vars=["Unproductive area"], y_vars=["Agricultural area"], hue="Elevation")
plt.title('Agricultural area vs. Unproductive area')
plt.show()
```
